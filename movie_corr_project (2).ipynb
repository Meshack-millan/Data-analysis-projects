```python
#import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read in the data
df=pd.read_csv(r'C:\Users\Administrator\Downloads\archive (4)\movies.csv')
df.head()
```

```python
#checking number of missing values
df.isnull().sum()
```

```python
pd.set_option('display.max_rows', None)
df.head(20)
```

```python
#getting the summary of the dataframe
df.describe()
```

```python
df=df.dropna(subset=['rating','released','score','votes','writer','country','company','gross','runtime','gross'])
df.isnull().sum()
```

```python
df['budget']=df['budget'].fillna(df['budget'].mean())
df.isnull().sum()
```

```python
#check the datatypes of the columns
df.dtypes
```

```python
df.head()
```

```python
#change data type of columns from float to integers to remove decimal point
df[['budget','gross','runtime','votes']]=df[['budget','gross','runtime','votes']].astype('int64')
```

```python
df.head()
```

```python
df['released_clean'] = df['released'].str.replace(r"\s*\(.*\)", "", regex=True)
df['released_clean'] = pd.to_datetime(df['released_clean'], errors='coerce')
```

```python
df.head()
```

```python
#drop the released column
df=df.drop(columns='released')
df.head()
```

```python
df['year']=pd.to_datetime(df['year'])
df.dtypes
```

```python
#create correct year column
df['year_correct']=df['released_clean'].astype('str').str[:4]
df.head()
```

```python
df=df.drop(columns=['year'])
df.head()
```

```python
#check the duplicates
df.duplicated().sum()
```

```python
#arrange in the dataframe in descending order in respect to the gross earnings
df.sort_values(by=['gross'],inplace=True,ascending=False)
df.head()
```

```python
#my hypothesis is that both the budget and the company have strong effect on the gross income generated here
#budget will have high correlation
#company high  correlation
```

```python
#check how budget is correlated wth gross
sns.scatterplot(x=df['budget'],y=df['gross'],color='red')
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.grid()
plt.show()
```

```python
sns.regplot(x=df['budget'],y=df['gross'],scatter_kws={'color':'black'},line_kws={'color':'blue'})
plt.show()
```

```python
#check the exact correlation figure
```

```python
df['budget'].corr(df['gross'])
```

```python
correlation_matrix=df.corr(numeric_only=True,method='pearson')
```

```python
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation_Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix=df_numerized.drop(columns=['name','released_clean','year_correct']).corr(method='pearson')
plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values(ascending=False)
```

```python
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs>0.5)]
high_corr
```

```python
#votes and budget have the highest correlation to gross earning
#company has low correlation
```
